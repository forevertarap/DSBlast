```python
# -*- coding: utf-8 -*-
"""
homework for EDA 1

·         Load data into a dataframe

·         Get column names

·         Get data types

·         Compute percentage of missing data of each column. 
          Which columns have high percentage of missing values

·         Print out first 25 records

·         Check if there are duplicated records. If yes, drop them.

·         Sort the dataframe based on Fiscal Year

·         Convert a string column (Base Salary) to numeric/float
"""

import pandas as pd
#import numpy as np

# load data into a dataframe
payroll_df = pd.read_csv('Citywide_Payroll_Data__Fiscal_Year_.csv',dtype='str')



# get column names
print('The column names are:\n', payroll_df.columns.tolist())



# Convert the 'Fiscal Year' field from a string to an int
payroll_df['Fiscal Year'] = payroll_df['Fiscal Year'].astype('int64')
# Convert the fields 'Regular Hours','OT Hours' from string to float
payroll_df[['Regular Hours','OT Hours']] = payroll_df[['Regular Hours','OT Hours']].astype('float64')

# get data types
print('\nThe data type for each column is:\n', payroll_df.dtypes)


# compute no. of missing data of each column
nan_df = pd.DataFrame(payroll_df.isnull().sum(),columns=['nan'])
# Compute percentage of missing data of each column.
nan_df['pct']=nan_df['nan']*100/len(payroll_df.index)

print('\nThe percentage of missing data of each column is (%):\n', nan_df['pct'])



# Which columns have high percentage of missing values.
flt_data = nan_df[nan_df['pct']>20]
print ('\nThe columns with high percentage (>20%) of missing values are:\n',flt_data)

print('\nThe column with the highest percentage of missing values is: '+ nan_df.idxmax()['pct'])
print('The highest percentage is: {}%'.format((nan_df.max()['pct'])))



# Print out first 25 records
print('\nThe first 25 records are:\n', payroll_df.head(25))



#  Check if there are duplicated records. If yes, drop them.
dropdup_df = payroll_df.drop_duplicates()
print('\nThe shape of original records is: ',payroll_df.shape)
print('The shape with the duplicated dropped is: ', dropdup_df.shape)

```

```python


# Sort the dataframe based on Fiscal Year
dropdup_df.loc[:,'Last Name'] = dropdup_df['Last Name'].str.strip()
sort_df = dropdup_df.sort_values(by=['Fiscal Year','Last Name'],ascending=[True,True])
print('\nThe last 25 records after sorting data by fiscal year and last name:\n',sort_df.tail(25)[sort_df.columns[:5]])


```

```python


# Convert a string column (Base Salary) to numeric/float
dropdup_df.loc[:,'Base Salary'] = dropdup_df['Base Salary'].str.extract('(\d+\.\d+)', expand=False).astype(float)
print('\nThe records after converting base salary to numeric are:\n',dropdup_df[['Last Name','Base Salary']].head(25))
```
